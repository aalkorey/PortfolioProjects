```python
@typecheck
def filter_for_apts(lopd: List[PropertyData]) -> List[PropertyData]:
    """
    Given an input list of property data, filter for property data that is only for apartments
    """
    # return []  # stub
    
    # Template based on arbitrary-sized and reference rule
    # Filtered list of property data for pt1
    acc = []       #type: List[PropertyData]
    for pd in lopd:
        if pd.type == PropertyType.APT:
            acc.append(pd)
    
    return acc
    

# Examples and tests for filter_for_apts
start_testing()
expect(filter_for_apts([PropertyData(Neighbourhood.WE, PropertyType.APT, 130.00),
                       PropertyData(Neighbourhood.Z, PropertyType.HOU, 80.00),
                       PropertyData(Neighbourhood.CW, PropertyType.LOF, 75.00)]), [PropertyData(Neighbourhood.WE, PropertyType.APT, 130.00)])
expect(filter_for_apts([PropertyData(Neighbourhood.IZ, PropertyType.HOU, 275.00),
                       PropertyData(Neighbourhood.OO, PropertyType.APT, 140.00),
                       PropertyData(Neighbourhood.BZ, PropertyType.APT, 75.00)]), [PropertyData(Neighbourhood.OO, PropertyType.APT, 140.00),
                                                                                   PropertyData(Neighbourhood.BZ, PropertyType.APT, 75.00)])
summary()
```

```python
@typecheck
def neighbourhood_in_list(lon: List[Neighbourhood], n: Neighbourhood) -> bool:
    """
    Return True if Neighbourhood is in input list. False otherwise
    """
    # return True  # stub
    if n in lon:
        return True
    else:
        return False
        
start_testing()
expect(neighbourhood_in_list([Neighbourhood.IZ, Neighbourhood.OO, Neighbourhood.BZ], Neighbourhood.IZ), True)
expect(neighbourhood_in_list([Neighbourhood.IZ, Neighbourhood.OO, Neighbourhood.BZ], Neighbourhood.DBOW), False)
summary()
```

```python
@typecheck
def convert_nei_to_strings(lon: List[Neighbourhood]) -> List[str]:
    """
    Given a list of neighbourhoods, convert it to a list of strings where each string matches its associated neighbourhood
    """
    # return []  # stub
    # Template based on arbitrary-sized and reference rule
    # List of strings
    acc = []       #type: List[str]
    for n in lon:
        # Template taken from Neighbourhood
        if n == Neighbourhood.BC:
            acc.append("Bijlmer-Centrum")
        elif n == Neighbourhood.BL:
            acc.append("Bos en Lommer")
        elif n == Neighbourhood.BO:
            acc.append("Bijlmer-Oost")
        elif n == Neighbourhood.BZ:
            acc.append("Buitenveldert - Zuidas")
        elif n == Neighbourhood.CO:
            acc.append("Centrum-Oost")
        elif n == Neighbourhood.CW:
            acc.append("Centrum-West")
        elif n == Neighbourhood.DANS:
            acc.append("De Aker - Nieuw Sloten")
        elif n == Neighbourhood.DBOW:
            acc.append("De Baarsjes - Oud-West")
        elif n == Neighbourhood.DPR:
            acc.append("De Pijp - Rivierenbuurt")
        elif n == Neighbourhood.GD:
            acc.append("Gaasperdam - Driemond")
        elif n == Neighbourhood.GS:
            acc.append("Geuzenveld - Slotermeer")
        elif n == Neighbourhood.IZ:
            acc.append("IJburg - Zeeburgereiland")
        elif n == Neighbourhood.NO:
            acc.append("Noord-Oost")
        elif n == Neighbourhood.NW:
            acc.append("Noord-West")
        elif n == Neighbourhood.OHIB:
            acc.append("Oostelijk Havengebied - Indische Buurt")
        elif n == Neighbourhood.O:
            acc.append("Osdorp")
        elif n == Neighbourhood.ON:
            acc.append("Oud-Noord")
        elif n == Neighbourhood.OO:
            acc.append("Oud-Oost")
        elif n == Neighbourhood.S:
            acc.append("Slotervaart")
        elif n == Neighbourhood.WA:
            acc.append("Watergraafsmeer")
        elif n == Neighbourhood.WE:
            acc.append("Westerpark")
        elif n == Neighbourhood.Z:
            acc.append("Zuid")
    
    return acc

start_testing()
expect(convert_nei_to_strings([Neighbourhood.IZ, Neighbourhood.OO, Neighbourhood.BZ]), ["IJburg - Zeeburgereiland", "Oud-Oost", "Buitenveldert - Zuidas"])
expect(convert_nei_to_strings([Neighbourhood.DBOW, Neighbourhood.CW, Neighbourhood.BZ]), ["De Baarsjes - Oud-West", "Centrum-West", "Buitenveldert - Zuidas"])
summary()
```

```python
@typecheck
def get_neis(lopd: List[PropertyData]) -> List[str]:
    """
    Given a list of property data, return a list of the unqiue neighbourhoods found in that input property data list
    """
    # return []  # stub
    # Template based on arbitrary-sized and reference rule
    # List of unique Neighbourhoods
    acc = []       #type: List[Neighbourhood]
    for pd in lopd:
        if neighbourhood_in_list(acc, pd.neighbourhood) == False:
            acc.append(pd.neighbourhood)
    
    final_acc = convert_nei_to_strings(acc)
    
    return final_acc

start_testing()
expect(get_neis([PropertyData(Neighbourhood.IZ, PropertyType.HOU, 275.00),
                 PropertyData(Neighbourhood.OO, PropertyType.APT, 140.00),
                 PropertyData(Neighbourhood.BZ, PropertyType.APT, 75.00)]), ["IJburg - Zeeburgereiland", "Oud-Oost", "Buitenveldert - Zuidas"])
expect(get_neis([PropertyData(Neighbourhood.DBOW, PropertyType.BOA, 180.00),
                 PropertyData(Neighbourhood.CW, PropertyType.BNB, 45.00),
                 PropertyData(Neighbourhood.DBOW, PropertyType.CRV, 75.00)]), ["De Baarsjes - Oud-West", "Centrum-West"])
summary()
```

```python
@typecheck
def get_sum(lof: List[float]) -> float:
    """
    Given an input list of floats, return the sum of the floats
    """
    # return 0  # stub
    sum = 0
    for f in lof:
        sum += f
    return sum

start_testing()
expect(get_sum([3.5, 8.8, 4.9]), 17.2)
expect(get_sum([5.5, 9.2, 6.3]), 21)
summary()
```

```python
@typecheck
def get_len(lof: List[float]) -> int:
    """
    Given an input list of floats, return the length of the list
    """
    # return 0  # stub
    length = 0
    for f in lof:
        length +=1
    return length

start_testing()
expect(get_len([4.6, 8.9, 3.3, 9.2]), 4)
expect(get_len([3.3, 2.2]), 2)
summary()
```

```python
@typecheck
def get_mean_prices(lopd: List[PropertyData]) -> List[float]:
    """
    Given a list of property data, return a list of mean prices per neighbourhood for all the neighbourhoods found in the input list of property data
    """
    # return []  # stub
    # Template based on arbitrary-sized and reference rule
    # List of unique Neighbourhoods
    neighbourhoods = []       #type: List[Neighbourhood]
    for pd in lopd:
        if neighbourhood_in_list(neighbourhoods, pd.neighbourhood) == False:
            neighbourhoods.append(pd.neighbourhood)
    
    mean_prices = []
    for n in neighbourhoods:
        prices_for_each_nei = []
        for pd in lopd:
            if pd.neighbourhood == n:
                prices_for_each_nei.append(pd.price)
        mean = get_sum(prices_for_each_nei)/get_len(prices_for_each_nei)
        mean_prices.append(mean)
    
    return mean_prices

# Examples and tests for get_apt_prices
start_testing()
expect(get_mean_prices([PropertyData(Neighbourhood.WE, PropertyType.APT, 130.00),
                        PropertyData(Neighbourhood.WE, PropertyType.APT, 59.00),
                        PropertyData(Neighbourhood.ON, PropertyType.APT, 99.00),
                        PropertyData(Neighbourhood.ON, PropertyType.APT, 55.00)]), [94.5, 77])
expect(get_mean_prices([PropertyData(Neighbourhood.DPR, PropertyType.APT, 80.00),
                        PropertyData(Neighbourhood.BL, PropertyType.APT, 79.00),
                        PropertyData(Neighbourhood.WA, PropertyType.APT, 99.00)]), [80, 79, 99])
summary()
```

```python
@typecheck
def find_max_price(lof: List[float]) -> float:
    """
    Given a list of floats, return the maximum value in the list
    """
    # return 0  # stub
    max_value = 0
    for f in lof:
        if f > max_value:
            max_value = f
    return max_value

start_testing()
expect(find_max_price([24.5, 78.9, 33.3]), 78.9)
expect(find_max_price([24.5, 78.9, 93.3]), 93.3)
summary()
```

```python
@typecheck
def analyze(lopd: List[PropertyData]) -> None: 
    """ 
    Given a list of property data, produces a cluster bar chart to show the average nightly price of apartments and houses for every neighbourhood found
    in the original list of property data
    """ 
    # Filter the properties that are not apartments or houses
    apts = filter_for_apts(lopd)
    houses = filter_for_houses(lopd)
    
    # Template based on visualization
    avg_prices_apts = get_mean_prices(apts)
    neighbourhood_apts = get_neis(apts)
    
    avg_prices_hou = get_mean_prices(houses)
    neighbourhood_hou = get_neis(houses)
    
    # Find max prices for labelling purposes in the plot
    max_apt_price = round(find_max_price(avg_prices_apts), 2)
    max_hou_price = round(find_max_price(avg_prices_hou), 2)
    
    # Set figure size
    fig = plt.figure()
    fig.set_figwidth(17)
    fig.set_figheight(11)
    
    # Make subplots with title and axes labels
    gs = fig.add_gridspec(2, hspace = 0)
    axs = gs.subplots(sharex = True, sharey = True)
    fig.suptitle("Prices of Apartments and Houses per Neighbourhood in Amsterdam", weight = "bold", fontsize = 20)
    fig.supylabel("Neighbourhoods", rotation = 0, weight = "bold")
    plt.xlabel("Average Nightly Price", weight = "bold")
    
    # Plot first cluster data
    plt.subplot(2, 1, 1)
    plt.barh(y = neighbourhood_apts, width = avg_prices_apts, height = 0.4, label = "Apartments", color = "red")
    plt.axvline(x = max_apt_price, linestyle = '--', color = 'black')
    plt.axvline(x = max_hou_price, linestyle = '--', color = 'black')
    plt.text(max_apt_price, 0.5, max_apt_price, fontsize = 'small', weight = "bold")
    plt.legend(loc = "upper center")
    
    # Plot second cluster data
    plt.subplot(2, 1, 2)
    plt.barh(y = neighbourhood_hou, width = avg_prices_hou, height = 0.4, label = "Houses")
    plt.axvline(x = max_apt_price, linestyle = '--', color = 'black')
    plt.axvline(x = max_hou_price, linestyle = '--', color = 'black')
    plt.text(max_hou_price, 0.5, max_hou_price, fontsize = 'small', weight = "bold")
    plt.legend(loc = "upper center")
    
    # Show plot
    plt.show()
    
    return None

@typecheck
def main(filename: str) -> None:
    """
    Reads the file from given filename, analyzes the data, returns the result (the plot)
    """
    # Template from HtDAP, based on function composition 
    return analyze(read(filename))


# Examples and tests for analyze 
start_testing()
expect(analyze(read("small_test_1.csv")), None) 
expect(analyze(read("small_test_2.csv")), None)
summary()

# Examples and tests for main
start_testing()
expect(main("small_test_1.csv"), None)
expect(main("small_test_2.csv"), None)
summary()
```
